```python
def condition_valid(ti,i,j):

    """  """

    di_j = distance_matrix[i,j]
    di = data_client_index['d'].iloc[i]
    dj = data_client_index['d'].iloc[j]
    ai = data_client_index['a'].iloc[i]
    bi = data_client_index['b'].iloc[i]
    aj = data_client_index['a'].iloc[j]
    bj = data_client_index['b'].iloc[j]

    tj = ti + di + di_j/V_moy

    return int(ti >= ai and ti<= bi-di and tj >= aj and tj<= bj-dj), tj 


def condition_valid_depot(j):

    # condition de validité de 0-->j
    d0_j = data_depot['DISTANCE_KM'].iloc[j]
    dj = data_client_index['d'].iloc[j]
    aj = data_client_index['a'].iloc[j]
    bj = data_client_index['b'].iloc[j]
    tj = 480.0+dj + d0_j/V_moy

    return  int(tj >= aj and tj<= bj-dj) , tj
```

```python
import random

def get_random_1_index(array,exclude_indices):
    # Trouver les indices des 1 dans le tableau, en excluant les indices donnés en argument
    ones_indices = [i for i, val in enumerate(array) if val == 1 and i not in exclude_indices ]
    
    # Vérifier s'il y a au moins un 1 dans le tableau après exclusion
    if len(ones_indices) == 0:
        return None

    # Choisir aléatoirement un indice parmi ceux des 1
    random_index = random.choice(ones_indices)

    # Déterminer l'ordre du 1 choisi
    #order = ones_indices.index(random_index) + 1
    order = sum(array[:random_index+1])
    # Retourner l'indice et l'ordre
    return random_index, order
```

```python
def generate_solution(reserved_list) : 

    # Création de la matrice P0,i

    P0i = np.array([condition_valid_depot(i)[0] for i in range(data_depot.shape[0])])
    i,order = get_random_1_index(P0i,reserved_list)

    sequences = []
    solution1 = []
    matrice_solution = []

    t = condition_valid_depot(i)[1] # ti
    k = 0

    sequences.append(order)
    solution1.append(i)
    P=P0i

    matrice_solution.append(list(P))

    while True :
        
        k = k+1
        #P = np.array([int(bool(condition_valid(t,i,j)[0]) and j not in solution1) for j in range(data_depot.shape[0])])
        P = np.array([int(bool(condition_valid(t,i,j)[0])) for j in range(data_depot.shape[0])])

        if P.sum() == 0 : 
            break
        
        if get_random_1_index(P,solution1+reserved_list)==None : 
            break
        
        j,order = get_random_1_index(P,solution1+reserved_list)

        sequences.append(order)
        solution1.append(j)
        
        t = condition_valid(t,i,j)[1] # tj = ti + cij

        i = j

        matrice_solution.append(list(P))

    matrice_solution = np.array(matrice_solution).T

    
    return {'codage' : sequences , 'solution' : solution1 , 'matrice_solution' : matrice_solution}
```

```python
def generate_cij(v_moy):

    C = np.zeros((distance_matrix.shape[0],distance_matrix.shape[0]))

    for i in range(distance_matrix.shape[0]):
        for j in range(distance_matrix.shape[0]):
            di = data_client_index['d'].iloc[i]
            dij = distance_matrix[i][j]
            C[i,j] = di + dij/v_moy

    return C



def generate_c0j(v_moy):

    C = np.zeros((distance_matrix.shape[0],1))

    for j in range(distance_matrix.shape[0]):

            d0_j = data_depot['DISTANCE_KM'].iloc[j]
            dj = data_client_index['d'].iloc[j]
            C[j] = 480+ dj + d0_j/v_moy

    return C
```

```python
def generate_global_solution():

    reserved_list = []
    all_clients = set(data_client_index.index)
    global_solution = []

    K = 0

    while True : 

        solution = generate_solution(reserved_list)['solution']
        global_solution.append(solution)
        reserved_list = reserved_list+solution

        K = K+1
        if len(reserved_list) == len(all_clients) : 
            break

    
    return global_solution
   
    
```

```python
def cout_fonction(X):

    K = len(X)
    
    cout = 0

    for Route in X : 

        c = 0
        for i in range(len(Route)-1):

            c  = c + Cij[Route[i],Route[i+1]]
        
        cout = cout+c

    return K + cout
```

```python
def valid_condition(route):

    """  """
    bool = True
    T = [condition_valid_depot(route[0])[1]]

    for i in range(len(route[1:])) : 
        T.append(condition_valid(T[i-1],route[i-1],route[i])[1])
        bool = bool*condition_valid(T[i-1],route[i-1],route[i])[0]


    return bool
```

```python
def permutations_ij(route):
    """
    Retourne toutes les permutations des éléments de position i et j pour i<j dans une liste.

    Args:
        route: Une liste d'éléments.

    Returns:
        Une liste de toutes les permutations possibles des éléments de position i et j pour i<j.
    """
    permutations = []
    for i in range(len(route)):
        for j in range(i+1, len(route)):
            permuted = route.copy()
            permuted[i], permuted[j] = permuted[j], permuted[i]
            permutations.append(permuted)
    return permutations
```

```python
def generate_initial_state(states):
    """Génère un état initial aléatoire."""
    return random.choice(states)
```

```python
def permutation_mutation(solution):
    gene1, gene2 = random.sample(range(len(solution)), 2)
    solution[gene1], solution[gene2] = solution[gene2], solution[gene1]
    return solution
```

```python
import math
def recuit_simule(initial_state,  temperature_initiale=1.0, temperature_finale=1e-8, alpha=0.99):

    """Implémente l'algorithme de recuit simulé."""

    history = []
    current_state = initial_state
    current_energy = cout_fonction(current_state)
    best_state = current_state
    best_energy = current_energy
    temperature = temperature_initiale

    while temperature > temperature_finale:
        
        #V = Voisinnage(current_state)
        new_state =  generate_global_solution()
        new_energy = cout_fonction(new_state)
        delta_energy = new_energy - current_energy

        if delta_energy < 0 or math.exp(-delta_energy / temperature) > random.random():
            current_state = new_state
            current_energy = new_energy

        if current_energy < best_energy:
            best_state = current_state
            best_energy = current_energy

        temperature *= alpha

        history.append(best_energy)

        print('best_energy : '  , best_energy)
        print('temperature : '  , temperature)
        print('new state : '  , current_state)


    return best_state, best_energy , history
```

### Genetique

```python
def select_elite_indices(fitness_scores, elite_size):
    """
    Sélectionne les indices des élites dans une liste de scores.

    :param fitness_scores: une liste de scores
    :param elite_size: le nombre d'élites à sélectionner
    :return: une liste d'indices correspondant aux élites
    """
    # Trier les scores en ordre décroissant
    sorted_indices = sorted(range(len(fitness_scores)), key=lambda i: fitness_scores[i], reverse=True)
    # Sélectionner les indices des élites
    elite_indices = sorted_indices[:elite_size]
    return elite_indices

def select_parents(routes, fitness_scores, num_parents):
    """
    Sélectionne des parents dans une population de routes en utilisant la sélection par rang.

    :param routes: une liste de routes
    :param fitness_scores: une liste de scores correspondant aux routes
    :param num_parents: le nombre de parents à sélectionner
    :return: une liste de paires de routes correspondant aux parents sélectionnés
    """
    # Trier les indices des routes par ordre décroissant de score
    sorted_indices = sorted(range(len(fitness_scores)), key=lambda i: fitness_scores[i], reverse=True)
    # Sélectionner les indices des parents
    parent_indices = sorted_indices[:num_parents]
    # Créer des paires de parents en sélectionnant des indices aléatoires dans la liste des parents
    parents = []
    for i in range(num_parents):
        j, k = random.sample(parent_indices, 2)
        parents.append((routes[j], routes[k]))
    return parents

```

```python
def roulette_selection(population, fitness_scores):
    max_fitness = sum(fitness_scores)
    selection_probs = [fitness/max_fitness for fitness in fitness_scores]
    return random.choices(population, weights=selection_probs, k=2)

def tournament_selection(population, fitness_scores):
    k = int(len(population)/10)
    if k < 2:
        k = 2
    tournament = random.sample(range(len(population)), k)
    tournament_fitness_scores = [fitness_scores[i] for i in tournament]
    selected_index = tournament_fitness_scores.index(min(tournament_fitness_scores))
    return [population[tournament[selected_index]], population[tournament[(selected_index+1)%k]]]


def order_crossover(parent1, parent2):
    gene1, gene2 = random.sample(range(len(parent1)), 2)
    if gene1 > gene2:
        gene1, gene2 = gene2, gene1

    child = [-1] * len(parent1)
    child[gene1:gene2+1] = parent1[gene1:gene2+1]
    remaining_genes = [g for g in parent2 if g not in child]

    index = gene2+1
    for i in range(len(remaining_genes)):
        if index == len(parent1):
            index = 0
        if child[index] == -1:
            child[index] = remaining_genes.pop(0)
        index += 1

    return child
```

```python
import random

def genetic_algorithm(routes, population_size=100, elite_size=10, mutation_rate=0.01, generations=100):
    """
    Résoud le problème de permutation de routes à l'aide de l'algorithme génétique.

    :param routes: une liste de listes
    :param population_size: taille de la population
    :param elite_size: taille de la population élite
    :param mutation_rate: taux de mutation
    :param generations: nombre de générations
    :return: la meilleure solution trouvée
    """
    # Créer la population initiale
    population = []
    for _ in range(population_size):
        individual = []
        for r in routes:
            if valid_condition(list(random.sample(r,len(r)))):
                individual.append(list(random.sample(r,len(r))))

        population.append(individual)

    # Évaluer la population
    fitness_scores = [1- cout_fonction(individual) for individual in population]

    # Boucle principale de l'algorithme génétique
    for _ in range(generations):
        # Sélectionner la population élite

        elite_indices = select_elite_indices(fitness_scores, elite_size)
        elite_population = [population[i] for i in elite_indices]

        # Générer la nouvelle population
        new_population = []
        child = []
        while len(new_population) < population_size - elite_size:
            parents = select_parents(population, fitness_scores, 2)

            parent1 = [parents[0][i] for i in range(len(parents[0]))]
            parent2 = [parents[1][i] for i in range(len(parents[1]))]
            
            child = order_crossover(parent1, parent2)
            child = permutation_mutation(child)
            new_population.append(child)

        # Ajouter la population élite à la nouvelle population
        new_population.extend(elite_population)
        print(child)
        # Évaluer la nouvelle population
        new_fitness_scores = [1- cout_fonction(individual) for individual in new_population[0]]

        # Remplacer la population par la nouvelle population
        population = new_population
        fitness_scores = new_fitness_scores

    # Trouver la meilleure solution
    best_index = fitness_scores.index(max(fitness_scores))
    best_solution = population[best_index]
    return best_solution
```

```python
def permute(i,j , L):
    A = L.copy()
    A[i-1],A[j-1]=A[j-1],A[i-1]
    return A
```

```python
def genere_voisinnage(order):
    n = len(order)
    V=[]
    for i in range(n):
        for j in range(i):
                order_optimal = permute(i,j , order)
                V.append(order_optimal)

    return V
```
